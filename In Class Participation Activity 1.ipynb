# Comparing Types of Encoding
| Code Type | Definition |
| --------- | -----------|
| Temporal | spike timing in relation to a stimulus |
| Rate | frequency of action potentials |
| Latency | delay between stimulus onset and first action potential |
| Synchronous | coincident spikes in different neurons to measure coordination |
| Population | combining information from multiple different neurons |

- *Tuning Curves:*
  - display how the firiing rate of a single neuron changes in response to different variations of a stimulus
  - stimulus feature vs. firing rate
